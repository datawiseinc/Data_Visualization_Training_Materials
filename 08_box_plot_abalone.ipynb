```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.offline as pyo
```

```python
df=pd.read_csv('data/abalone.csv')
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
data = [go.Histogram(x=df['length'],xbins=dict(size=0.02), name='Histogram_length') ]
```

```python
layout = go.Layout(title='Histogram of Abalone Length', xaxis=dict(title='length'), yaxis=dict(title='count'))
```

```python
figure = go.Figure(data=data, layout=layout)
```

```python
pyo.plot(figure, filename='08_histogram_abalone_length.html')
```
